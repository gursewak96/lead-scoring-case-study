## Problem Statement

X Education has appointed you to help them select the most promising leads, i.e. the leads that are most likely to convert into paying customers. The company requires you to build a model wherein you need to assign a lead score to each of the leads such that the customers with a higher lead score have a higher conversion chance and the customers with a lower lead score have a lower conversion chance. The CEO, in particular, has given a ballpark of the target lead conversion rate to be around 80%

## Step 10: Feature Selection Using RFE

Since we have so many feature to deal with we cant go further with all feature intact for manual selection. Therefore using RFE to get the top 15 features

```python
# Getting top 15 features using RFE
log_reg = LogisticRegression()
rfe = RFE(log_reg,n_features_to_select=15)
rfe = rfe.fit(X_train, y_train)

list(zip(X_train.columns, rfe.support_, rfe.ranking_))
```

```python
#Get the rfe_supported columns name
rfe_support_columns = X_train.columns[rfe.support_]
rfe_support_columns
```

Lets assess the model with StatsModels

```python
X_train_sm = sm.add_constant(X_train[rfe_support_columns])
log_model2 = sm.GLM(y_train, X_train_sm, family=sm.families.Binomial())
res = log_model2.fit()
res.summary()
```

Lets create a function to predict target variable with cutoff of 0.5, if the convert probablity is above 0.5 then 1 (lead can be converted) else 0 (can't be converted)

```python
# define a function to predict target variable
def predict_target(res,X_train_sm):
    y_train_pred = res.predict(X_train_sm).values.reshape(-1)
    y_train_pred_final["Convert_prob"] = y_train_pred
    
    # cutoff is at 0.5 if convert_prob > 0.5 ==> 1 else 0
    y_train_pred_final['Predicted'] = y_train_pred_final.Convert_prob.map(lambda x: 1 if x > 0.5 else 0)
    return y_train_pred_final
```

Create dataframe with the true convertion and the predicted probabilities

```python
#create new column "predicted" with 1 if convert_prob > 0.5 else 0
y_train_pred_final = predict_target(res,X_train_sm)
y_train_pred_final
```

Lets get the accuracy score

```python
print("Accuracy Score: ", metrics.accuracy_score(y_train_pred_final.Convert, y_train_pred_final.Predicted))
```

Lets check the VIF

```python
#create function to check the VIF
def check_VIF(X_train):
    vif_df = pd.DataFrame()
    vif_df['Features'] = X_train.columns
    vif_df['Variance Inflation Factor'] = [variance_inflation_factor(X_train.values,i) for i in range(X_train.shape[1])]
    vif_df['Variance Inflation Factor'] = round(vif_df['Variance Inflation Factor'],2)
    vif_df = vif_df.sort_values(by= 'Variance Inflation Factor', ascending=False)
    return vif_df
    
vif_df = check_VIF(X_train[rfe_support_columns])
vif_df 
```

Every variable possesses a favorable VIF value. However, we noticed previously that the "Tags_invalid_number" column exhibits a high p-value, therefore we will eliminate this column and reconstruct the model.

```python
# drop the "Tags_invalid number" column
rfe_support_columns = rfe_support_columns.drop("Tags_invalid number")
rfe_support_columns
```

Lets define a function to evaluate the model


```python
def evaluate_model(y_true, y_predicted, print_score=False):
    confusion = metrics.confusion_matrix(y_true, y_predicted)
    
    #true positive [TP]
    TP = confusion[1,1]
    #true negative [TN]
    TN = confusion[0,0]
    #false positive 
    FP = confusion[0,1]
    #false negative
    FN = confusion[1,0]
    
    accuracy_sc = metrics.accuracy_score(y_true, y_predicted)
    sensitivity_score = TP/float(TP + FN)
    specificity_score = TN/float(TN + FP)
    precision_sc = precision_score(y_true, y_predicted)
    
    
    if print_score:
        print("Confusion Matrix: \n", confusion)
        print("Accuracy: ",accuracy_sc)
        print("Sensitivity: ",sensitivity_score)
        print("Specificity: ",specificity_score)
        print("Precision: ", precision_sc)
        
    return accuracy_sc, sensitivity_score, specificity_score, precision_sc
```

```python
#evalatue the model
evaluate_model(y_train_pred_final.Convert, y_train_pred_final.Predicted, print_score=True)
```

In our case, sensitivity holds utmost significance as it enables us to predict a high number of leads, thereby enhancing our productivity and revenue in a compelling manner. Simply put, sensitivity is the most important metric for us because it helps us predict as many high-quality leads as possible. This, in turn, boosts our productivity and revenue for the company that is X Education.
